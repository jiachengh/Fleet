```python
import numpy as np
import matplotlib.pyplot as plt
import math

np.set_printoptions(threshold=np.inf)


def avg(x):
    return sum(x) / float(len(x))

def percentile(data, p, ifsorted=False):
    if ifsorted:
        s = data
    else:
        s = sorted(data)
    index = math.ceil(len(s) * p) - 1
    return s[index]
```
